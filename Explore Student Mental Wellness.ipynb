```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/kaggle/input/student-mental-health/Student Mental health.csv")
```

```python
df.head()
```

# Gender Distribution - Visualize the distribution of genders among students

```python
# Countplot for gender distribution
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Choose your gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

# Age Distribution - Visualize the distribution of ages among students

```python
# Histogram for age distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# Course Distribution - Visualize the distribution of courses among students

```python
# Countplot for course distribution
plt.figure(figsize=(10, 10))
sns.countplot(data=df, y='What is your course?')
plt.title('Course Distribution')
plt.xlabel('Count')
plt.ylabel('Course')
plt.show()
```

# CGPA Distribution - Visualize the distribution of CGPA among students

```python
# Stripplot for CGPA distribution
plt.figure(figsize=(8, 6))
sns.stripplot(data=df, y='What is your CGPA?')
plt.title('CGPA Distribution')
plt.ylabel('CGPA')
plt.show()
```

# Marital Status Distribution - Visualize the distribution of marital status among students

```python
# Countplot for marital status distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Marital status')

# Get the count for each category
counts = df['Marital status'].value_counts()

for i, count in enumerate(counts):
    plt.text(i, count, f'{count:.0f}', ha='center', va='bottom')

plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

# Depression Status - Visualize the distribution of students with depression

```python
# Countplot for depression status
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Do you have Depression?')
plt.title('Depression Status')
plt.xlabel('Depression')
plt.ylabel('Count')

# Add exact numbers on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.show()
```
